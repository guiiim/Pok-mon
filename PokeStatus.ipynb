```python
#import bibliotecas utilizadas
import pokebase as pb
from pokebase import pokemon
import pandas as pd
```

```python
geracoes = [1, 2, 3, 4, 5, 6, 7, 8]
names = ["Generation", "ID", "Name", "Type1", "Type2", "HP", "Defense", "Attack", "SP Defense", "Link"]
dfFotos = pd.DataFrame(columns = names)
geracoes = [1]
dados = []
```

```python
#cria os arrays
geracoes = [] #array para registrar todas as gerações existentes
names = ["Generation", "ID", "Name", "Type1", "Type2", "HP", "Defense", "Attack", "SP Defense", "Link"] #array com o nome das colunas para o dataframe
dados = [] #array para armazenar os dados
```

```python
#verifica o número de gerações
for i in range(1, 1000):
    try: #testa as gerações a partir da 1ª
        gen_resource = pb.generation(i)
        geracoes.append(i)
    except: #quando a geração não existir, encerra o loop
        break
```

```python
for i in geracoes:
    gen_resource = pb.generation(i) #busca todos os dados da geração atual

    for nomes in gen_resource.pokemon_species: #pega todos os nomes da geração
        try:
            poke = pokemon(nomes.name.title().lower()) #busca os status do nome atual
            id = str(poke.id) #salva ID do pokémon
            name = nomes.name.title() #salva nome do pokémon
            type1 = poke.types[0].type.name #salva o tipo 1 do pokémon
            
            try: #verifica se o pokémon tem o tipo 2
                type2 = poke.types[1].type.name #se sim, salva o tipo 2 do pokémon
            except:
                type2 = None #se não, salva o tipo 2 do pokémon como "Nenhum"
                
            hp = poke.stats[5].base_stat #salva o hp do pokémon
            defense = poke.stats[3].base_stat #salva a defesa do pokémon
            attack = poke.stats[4].base_stat #salva o ataque do pokémon
            spDefense = poke.stats[1].base_stat #salva a defesa SP do pokémon
            link = poke.sprites.front_default #salva o link da imagem do pokémon

            dados.append([i, id, name, type1, type2, hp, defense, attack, spDefense, link]) #acrescenta os dados ao array
            
        except:
            pass
```

```python
#cria o dataframe com os dados do array
df = pd.DataFrame.from_records(dados, columns = names)
```
